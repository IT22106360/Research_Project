```python
import os
```

```python
# Folder containing PNG images
folder_path = r"D:\Y4 Research\datasets\Llava_claims_dataset\Images"

# Get only PNG files
images = [f for f in os.listdir(folder_path) if f.lower().endswith(".png")]

# Sort to keep consistent order
images.sort()

# Rename files
for idx, filename in enumerate(images, start=1):
    old_path = os.path.join(folder_path, filename)
    
    new_name = f"image_{idx:03d}.png"
    new_path = os.path.join(folder_path, new_name)
    
    os.rename(old_path, new_path)
    print(f"{filename} → {new_name}")
```
