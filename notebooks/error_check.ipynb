```python
import os
import re

# -----------------------------
# PATH
# -----------------------------
JSON_DIR = r"D:\Y4 Research\datasets\ocr_json\ing_nut_set1\seperate_DPI_removed"

# -----------------------------
# RENAME FILES
# -----------------------------
for filename in os.listdir(JSON_DIR):

    if not filename.lower().endswith(".json"):
        continue

    old_path = os.path.join(JSON_DIR, filename)

    # remove DPI suffix like .400, .1.400, .5.400 before .json
    new_filename = re.sub(r'\.\d+\.json$', '.json', filename)

    if new_filename == filename:
        continue  # no change needed

    new_path = os.path.join(JSON_DIR, new_filename)

    # avoid overwriting existing files
    if os.path.exists(new_path):
        print(f"âš  Skipped (already exists): {new_filename}")
        continue

    os.rename(old_path, new_path)
    print(f"âœ… Renamed: {filename} â†’ {new_filename}")

print("\nðŸŽ‰ DPI suffix removal completed.")
```

```python
base_path = r"D:\Y4 Research\datasets\annotation_dataset"

# Regex to match a trailing ".digits" (e.g., .400, .300)
pattern = re.compile(r"\.\d+$")

for folder_name in os.listdir(base_path):
    old_path = os.path.join(base_path, folder_name)

    # Process only directories
    if os.path.isdir(old_path):
        new_name = pattern.sub("", folder_name)

        # Rename only if a change is needed
        if new_name != folder_name:
            new_path = os.path.join(base_path, new_name)
            os.rename(old_path, new_path)
            print(f"Would rename: {folder_name} â†’ {new_name}")

print("Done.")
```

```python
import os
import shutil

# Paths
RAW_DIR = r"D:\Y4 Research\datasets\ingredient & nutrition Images\png_converted"
ANNOTATION_DIR = r"D:\Y4 Research\datasets\annotation_dataset - Copy"

# Get list of raw images
raw_images = os.listdir(RAW_DIR)

# Function to find best match for a folder base name
def find_raw_image(folder_base, raw_images):
    for img in raw_images:
        img_base = os.path.splitext(img)[0]
        # Remove possible DPI suffix like _2.400
        img_base_clean = img_base.split('_')[0] + ''.join(['_' + x for x in img_base.split('_')[1:] if not x.replace('.', '', 1).isdigit()])
        if folder_base.startswith(img_base_clean):
            return img
    return None

# Iterate over annotation folders
for folder_name in os.listdir(ANNOTATION_DIR):
    folder_path = os.path.join(ANNOTATION_DIR, folder_name)
    if os.path.isdir(folder_path):
        matched_raw_img = find_raw_image(folder_name, raw_images)
        if matched_raw_img:
            src_path = os.path.join(RAW_DIR, matched_raw_img)
            dst_path = os.path.join(folder_path, "raw_image.png")
            shutil.copy2(src_path, dst_path)  # Copy raw image into folder
            print(f"âœ… Copied {matched_raw_img} to {folder_name}/raw_image.png")
        else:
            print(f"âš  No matching raw image found for folder: {folder_name}")
```

```python
# Pattern: remove .digits ONLY if it appears before .png
dpi_pattern = re.compile(r"\.(\d+)(?=\.png$)", re.IGNORECASE)

for folder_name in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder_name)

    if not os.path.isdir(folder_path):
        continue

    for file_name in os.listdir(folder_path):
        if file_name.lower().endswith(".json"):
            json_path = os.path.join(folder_path, file_name)

            with open(json_path, "r", encoding="utf-8") as f:
                data = json.load(f)

            if "image_id" in data and isinstance(data["image_id"], str):
                original_id = data["image_id"]
                cleaned_id = dpi_pattern.sub("", original_id)

                if original_id != cleaned_id:
                    data["image_id"] = cleaned_id

                    with open(json_path, "w", encoding="utf-8") as f:
                        json.dump(data, f, indent=4, ensure_ascii=False)

                    print(
                        f"[UPDATED] {folder_name}: "
                        f"{original_id} â†’ {cleaned_id}"
                    )

            # One OCR JSON per folder
            break

print("Finished removing DPI suffixes from image_id.")
```
