```python
image_dir = r"D:\Y4 Research\datasets\ingredient & nutrition Images\grayscale_clahe_denoised"
output_json = r"D:\Y4 Research\datasets\ocr_json\ing_nut set1\tesseract_bounding_box_ocr.json"
```

```python
import os
```

```python
records = []

for img_name in tqdm(os.listdir(image_dir)):
    img_path = os.path.join(image_dir, img_name)

    image = cv2.imread(img_path)
    if image is None:
        continue

    ocr_text = pytesseract.image_to_string(
        image,
        config="--oem 3 --psm 6"
    )

    data = pytesseract.image_to_data(
        image,
        output_type=pytesseract.Output.DICT,
        config="--oem 3 --psm 6"
    )

    words = []
    for i in range(len(data["text"])):
        if data["text"][i].strip():
            words.append({
                "text": data["text"][i],
                "confidence": int(data["conf"][i]),
                "bbox": [
                    data["left"][i],
                    data["top"][i],
                    data["width"][i],
                    data["height"][i]
                ]
            })

    records.append({
        "image_id": img_name,
        "ocr_text": ocr_text,
        "words": words
    })
```

```python
final_json = {
    "ocr_engine": "tesseract",
    "total_images": len(records),
    "psm": 6,
    "oem": 3,
    "preprocessing": "already_done",
    "data": records
}
```

```python
os.makedirs(os.path.dirname(output_json), exist_ok=True)

with open(output_json, "w", encoding="utf-8") as f:
    json.dump(final_json, f, indent=2, ensure_ascii=False)
```

OCR for each image in set 1

```python
import os
import json
import cv2
import pytesseract
from tqdm import tqdm

# -----------------------------
# PATHS
# -----------------------------
INPUT_DIR = r"D:\Y4 Research\datasets\ingredient & nutrition Images\grayscale_clahe_denoised"
OUT_DIR = r"D:\Y4 Research\datasets\ocr_json\ing_nut_set1\seperate"

os.makedirs(OUT_DIR, exist_ok=True)

# -----------------------------
# Tesseract config
# -----------------------------
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
custom_config = r"--oem 3 --psm 6"

# -----------------------------
# OCR loop
# -----------------------------
for img_name in tqdm(os.listdir(INPUT_DIR)):
    if not img_name.lower().endswith((".png", ".jpg", ".jpeg")):
        continue

    img_path = os.path.join(INPUT_DIR, img_name)
    image = cv2.imread(img_path)

    if image is None:
        print("Failed to read:", img_path)
        continue

    # Full OCR text
    ocr_text = pytesseract.image_to_string(image, config=custom_config)

    # OCR with bounding boxes
    data = pytesseract.image_to_data(image, config=custom_config, output_type=pytesseract.Output.DICT)

    words = []
    n = len(data["text"])
    for i in range(n):
        text = data["text"][i].strip()
        if text == "":
            continue
        try:
            conf = int(float(data["conf"][i]))
        except:
            conf = 0
        word_info = {
            "text": text,
            "confidence": conf,
            "bbox": [
                int(data["left"][i]),
                int(data["top"][i]),
                int(data["width"][i]),
                int(data["height"][i])
            ]
        }
        words.append(word_info)

    output_json = {
        "image_id": img_name,
        "ocr_text": ocr_text,
        "words": words
    }

    out_path = os.path.join(OUT_DIR, os.path.splitext(img_name)[0] + ".json")
    with open(out_path, "w", encoding="utf-8") as f:
        json.dump(output_json, f, indent=2, ensure_ascii=False)
```

```python
import os
import json

RAW_IMG_DIR = r"D:\Y4 Research\datasets\ingredient & nutrition Images\raw"
OCR_DIR = r"D:\Y4 Research\datasets\ocr_json\ing_nut_set1\seperate"
OUT_FILE = r"D:\Y4 Research\datasets\ocr_json\ing_nut_set1\annotation ready\annotation_import.json"

tasks = []

for ocr_file in os.listdir(OCR_DIR):
    if not ocr_file.endswith(".json"):
        continue

    with open(os.path.join(OCR_DIR, ocr_file), "r", encoding="utf-8") as f:
        ocr_data = json.load(f)

    image_name = ocr_data["image_id"]

    task = {
        "data": {
            "image": os.path.join(RAW_IMG_DIR, image_name),
            "ocr": ocr_data["words"]
        }
    }
    tasks.append(task)

with open(OUT_FILE, "w", encoding="utf-8") as f:
    json.dump(tasks, f, indent=2)

print("Annotation dataset created:", OUT_FILE)
```

Annotations & Images dataset

```python
import os
import json
import shutil

# -----------------------------
# PATHS
# -----------------------------
RAW_DIR = r"D:\Y4 Research\datasets\ingredient & nutrition Images\png_converted"
OCR_DIR = r"D:\Y4 Research\datasets\ocr_json\ing_nut_set1\seperate_DPI_removed"
OUT_DIR = r"D:\Y4 Research\datasets\annotation_dataset"

os.makedirs(OUT_DIR, exist_ok=True)

# -----------------------------
# PROCESS EACH OCR FILE
# -----------------------------
for ocr_file in os.listdir(OCR_DIR):

    if not ocr_file.endswith(".json"):
        continue

    ocr_path = os.path.join(OCR_DIR, ocr_file)

    with open(ocr_path, "r", encoding="utf-8") as f:
        ocr_data = json.load(f)

    image_name = ocr_data["image_id"]
    image_base = os.path.splitext(image_name)[0]

    raw_image_path = os.path.join(RAW_DIR, image_name)

    if not os.path.exists(raw_image_path):
        print(f"⚠ Missing image: {image_name}")
        continue

    # Create item folder
    item_dir = os.path.join(OUT_DIR, image_base)
    os.makedirs(item_dir, exist_ok=True)

    # Copy raw image
    shutil.copy(
        raw_image_path,
        os.path.join(item_dir, "raw_image.png")
    )

    # Copy OCR JSON
    shutil.copy(
        ocr_path,
        os.path.join(item_dir, "ocr_result.json")
    )

print("✅ Dataset structure created successfully.")
```
